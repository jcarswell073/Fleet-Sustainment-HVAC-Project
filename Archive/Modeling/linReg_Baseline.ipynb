```python
import pandas as pd
data = pd.read_csv('dataset_je2.csv')
#Use to set datatypes
data['uic'] = data['uic'].astype('string')
data['apl'] = data['apl'].astype('string')
data['cdm_rin'] = data['cdm_rin'].astype('string')
data['csmp_narrative_summary'] = data['csmp_narrative_summary'].astype('string')
data['date_closing'] = pd.to_datetime(data['date_closing'], errors='coerce')
data['date_maintenance_action'] = pd.to_datetime(data['date_maintenance_action'], errors='coerce')
data['due_date'] = pd.to_datetime(data['due_date'], errors='coerce')
data['eic'] = data['eic'].astype('string')
data['equipment_nomenclature'] = data['equipment_nomenclature'].astype('string')
data['eswbs_opening'] = data['eswbs_opening'].astype('string')

data['iuc_screening_code'] = data['iuc_screening_code'].astype('string')
data['location'] = data['location'].astype('string')
data['safety_code'] = data['safety_code'].astype('string')
data['tycom_screening_code'] = data['tycom_screening_code'].astype('string')
data['type_availability_code'] = data['type_availability_code'].astype('int')
data['type_of_maintenance_action'] = data['type_of_maintenance_action'].astype('string')
data['work_center'] = data['work_center'].astype('string')
data['def_narrative'] = data['def_narrative'].astype('string')
data['nondef_narrative'] = data['nondef_narrative'].astype('string')
data['closing_narrative'] = data['closing_narrative'].astype('string')
data['ima_narrative'] = data['ima_narrative'].astype('string')
 
 
 
data['job_status'] = data['job_status'].astype('string')
data['priority'] = data['priority'].astype('string')

data['type_of_availability_needed'] = data['type_of_availability_needed'].astype(
    'string')
data['tycom_screening'] = data['tycom_screening'].astype('string')
data['when_discovered'] = data['when_discovered'].astype('string')
data['intermediate_unit_commander_screening'] = data['intermediate_unit_commander_screening'].astype(
    'string')
data['deferral_reason'] = data['deferral_reason'].astype('string')
data['opening_ship_system'] = data['opening_ship_system'].astype('string')

data['action_taken'] = data['action_taken'].astype('string')
data['cause'] = data['cause'].astype('string')
data['ship'] = data['ship'].astype('string')
data['issue_date'] = pd.to_datetime(data['issue_date'], errors='coerce')

data['issue_apl'] = data['issue_apl'].astype('string')

data['issue_eic'] = data['issue_eic'].astype('string')

data['niin_nomenclature'] = data['niin_nomenclature'].astype('string')

data['unit_of_issue'] = data['unit_of_issue'].astype('string')

data['source_code'] = data['source_code'].astype('string')


data['jsn'] = data['jsn'].astype('string')

data['demand_date'] = pd.to_datetime(data['demand_date'], errors='coerce')

data['date_closing_s'] = pd.to_datetime(data['date_closing_s'], errors='coerce')

#Put in features that were created

#dataset_je
#loc 1-4 string
data['loc1'] = data['loc1'].astype('string')
data['loc2'] = data['loc2'].astype('string')
data['loc3'] = data['loc3'].astype('string')
data['loc4'] = data['loc4'].astype('string')

data['action_taken_code'] = data['action_taken_code'].astype('string')

#cwtDays, totPrice, totMaterialCost float (dataset_je)
data['cwtDays'] = data['cwtDays'].astype('float')
data['totPrice'] = data['totPrice'].astype('float')
data['totMaterialCost'] = data['totMaterialCost'].astype('float')

data['daysOpen'] = data['daysOpen'].astype('float') #dataset_je1
```

```python
categorical_features = ['action_taken_code','cause_code','when_discovered_code','deferral_reason_code','loc1','status_code','eic','uic']
df = pd.get_dummies(data, columns=categorical_features,dtype = int)
df2 = df.filter(regex = 'job_seq|action_taken_code|cause_code|when_discovered_code|deferral_reason_code|loc1|daysOpen|status_code|eic|uic')
```

```python
df2 = df2.drop('uic_s',axis=1)
df2 = df2.drop('issue_eic',axis=1)
df2 = df2[df2['daysOpen'].notna()]
features = [col for col in df2.columns if col != 'daysOpen']
```

```python
X = df2[features]
y = df2['daysOpen']
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,y)
```

```python
r2_score = model.score(X,y) #0.22624042246389053
```
